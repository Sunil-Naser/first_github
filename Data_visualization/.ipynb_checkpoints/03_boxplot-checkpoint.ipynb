##  How we made boxplot (code of line change from other like lineplot,barplot)

```python
# Import libraries
import seaborn
# canvas (bloon board)
seaborn.set(style='whitegrid')
# Load dataset
boat= seaborn.load_dataset("titanic")
# draw a boxplot with
seaborn.boxplot(x="class",
                y="fare",
                data=boat)
```

## Add color pick from color site (hex color picker)

```python
# Import libraries
import seaborn as sns
# canvas (bloon board)
sns.set(style='whitegrid')
# Load dataset
tip= sns.load_dataset("tips")
tip
# draw a boxplot with color change(and this code color i spick from google (hex color picker)
sns.boxplot(x="day",y="tip",data=tip, color="#a83285")  
```

## Add saturation

```python
# Import libraries
import seaborn as sns
# canvas (bloon board)
sns.set(style='whitegrid')
# Load dataset
tip= sns.load_dataset("tips")
tip
# draw a boxplot with (here we use saturation for color contracrt
sns.boxplot(x="tip",y="day",data=tip, saturation=0.5,color="#a83285") 
```

## Some advance thing in boxplot like how we change size,color,add label ect

```python
import seaborn as sns 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

b
#how to change mark ,color and size in barplot
sns.boxplot(x="survived", y="age",showmeans=True,
            meanprops={"marker":"*",
                       "markersize":"10",
                       "markeredgecolor":"red"},data=boat)
# how we add label and change their size in barplot
plt.xlabel("how man people are survived", size=10,weight="bold"),
plt.ylabel("what the age of these people", size=10,weight="bold"),
plt.title("how many people are survived or sank?",size=20, weight="bold")
plt.show()
```

## What is Facet Grid?
- In which we have made multi graph with one data sheet 

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tip=sns.load_dataset("tips")
tip.head(10)
```

```python
tip=sns.load_dataset("tips")
tip.describe()
```

```python
fg=sns.FacetGrid(tip,col="sex",hue="day")
fg.map(plt.bar,"total_bill","tip").add_legend()
plt.show()
```

- If you want plot show days wise then what we do?

```python
fg=sns.FacetGrid(tip,col="day",hue="sex")
fg.map(plt.bar,"total_bill","tip").add_legend()
plt.show()
```
