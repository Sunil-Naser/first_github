## Import Libraries
Seaborn automatically install these libraries
- numpy
- scipy
- pandas
- matplotlip

```python
# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
flower= sns.load_dataset("iris")
flower
# draw a line plot
sns.lineplot(x="sepal_length",y="sepal_width", data=flower)
plt.show()
```

## Adding title

```python
# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
flower= sns.load_dataset("iris")
flower
# draw a line plot
sns.lineplot(x="sepal_length",y="sepal_width", data=flower)
plt.title("my first graph in jupyter note book")
plt.show()
```

## Adding limits

```python
# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
flower= sns.load_dataset("iris")
flower
# draw a line plot
sns.lineplot(x="sepal_length",y="sepal_width", data=flower)
plt.title("my first graph in jupyter note book")
plt.xlim(2)
plt.ylim(2)
plt.show()
```

## Siza of figure

```python
# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
flower= sns.load_dataset("iris")
# change figure size
plt.figure(figsize=(6,2))

# draw a line plot
sns.lineplot(x="sepal_length",y="sepal_width", data=flower)
plt.title("my first graph in jupyter note book")

plt.show()
```

```python
# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
flower= sns.load_dataset("iris")
flower
# draw a line plot
sns.lineplot(x="sepal_length",y="sepal_width", data=flower)
plt.title("my first graph in jupyter note book")
# Set style
sns.set_style("darkgrid")
sns.set_style(style=None, rc=None)
plt.show()
```
